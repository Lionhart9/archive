```python
import re

# Your input string
a = [' of funding available (Item 8.6 divided by item 8.3) 5.23 blah blah ']

# Define a regex pattern to find the text after "item 8.3)"
pattern = r'item 8\.3\)\s([^\s]*)'

# Find the match
match = re.search(pattern, a[0], re.IGNORECASE)

# Extract the first word/characters after "item 8.3)"
if match:
    result = match.group(1)
    print("Extracted text:", result)
else:
    print("Pattern not found")
```
